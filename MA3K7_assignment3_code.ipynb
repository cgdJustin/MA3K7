```python
import random

def simulate_paper_draw(n):
    # 初始化数字集合
    papers = list(range(1, n+1))
    
    # 当集合中数字数量大于1时，继续执行操作
    while len(papers) > 1:
        # 随机选择两个数字
        a, b = random.sample(papers, 2)
        
        # 计算差值，并更新集合
        papers.remove(a)
        papers.remove(b)
        papers.append(abs(a - b))
    
    # 返回最后剩下的数字
    return papers[0]

# 执行模拟程序
simulate_paper_draw(2022)
```

## Odd even case

```python
even,odd = 0,0
for i in range(1000):
    x = simulate_paper_draw(4)
    if x%2 == 0:    
        even += 1 
    else:
        odd += 1 
print(even,odd)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_paper_draw(n):

    papers = list(range(1, n+1))
    
    while len(papers) > 1:
        # randomly pick two numbers
        a, b = random.sample(papers, 2)
        
        # Calculating the difference
        papers.remove(a)
        papers.remove(b)
        papers.append(abs(a - b))
    
    # final number
    return papers[0]
```

```python
even, odd = 0, 0
# Track outcomes
outcomes = []
means = []
for i in range(1,1001):
    for j in range(1000):
        
        x = simulate_paper_draw(i)
        outcomes.append(x)
        
        
        if x % 2 == 0:    
            even += 1 
        else:
            odd += 1 
        
    means.append(np.mean(outcomes))
    outcomes = []
print(means)
```

```python
# Plotting the mean of the outcomes
import matplotlib.pyplot as plt

plt.plot(range(1, 1001), means, marker='o', linestyle='-', color='b')
plt.title('Mean of Final Numbers for Different n')
plt.xlabel('n ')
plt.ylabel('Mean of Final Numbers')
plt.grid(True)
plt.show()
```

```python

# Plotting the distribution
plt.hist(outcomes, bins=range(1, 2025), align='left', color='blue', edgecolor='black')
plt.title('Distribution of Final Numbers')
plt.xlabel('Final Number')
plt.ylabel('Frequency')
plt.xticks(range(1, 2025))
plt.show()
```
