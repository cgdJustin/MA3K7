### The following program will display the simulation of the coin throw game
In this case, we only consider would we ever hit 25

```python
import numpy as np

def simulate_coin_throws(N = 24,target=25, simulations = 10000):
    hits = 0
    for i in range(simulations):
        steps = 0
        for j in range(N):
            steps += np.random.choice([1, 2])
            if steps == target:
                hits += 1
                break
            if steps > target:
                break
        
    return hits / simulations
```

```python
simulate_coin_throws()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Fixed parameters
N = 24
target = 25
simulation_steps = [5, 10, 50, 100, 500, 1000, 10000,20000,30000,40000,100000,200000]  # Modify as needed

# Store results
probabilities = []

for simulations in simulation_steps:
    probability = simulate_coin_throws(N=N, target=target, simulations=simulations)
    probabilities.append(probability)
```

```python

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(simulation_steps, probabilities, marker='o', linestyle='-', color='b', label='Simulated Probability')
plt.axhline(y=0.6667, color='r', linestyle='--', label='Expected Probability (y = 0.6667)')
plt.title('Probability of Hitting Target with Varying Number of Simulations')
plt.xlabel('Number of Simulations')
plt.ylabel('Probability of Hitting Target')
plt.xscale('log')  # Optional: Log scale for x-axis to better visualize changes over a wide range of simulations
plt.legend()
plt.grid(True)
plt.show()

```

In this case, we only consider the final number

```python
import numpy as np

def simulate_coin_throws_final(n, target=25, simulations=10000):
    hits = 0
    for i in range(simulations):
        steps = 0
        for i in range(n):
            steps += np.random.choice([1, 2])
        if steps == target:
            hits += 1
                
    return hits / simulations
```

```python
import matplotlib.pyplot as plt

# 设置n的范围
n_values = range(10, 24)
# 计算每个n值对应的概率
probabilities1 = [simulate_coin_throws_final(n, simulations=10000) for n in n_values]
```

```python
sum(probabilities1)
```

```python
# 绘制概率分布图
plt.figure(figsize=(10, 6))
plt.plot(n_values, probabilities1, marker='o', linestyle='-', color='b')
plt.title('Probability of Landing on Step 25 for different N')
plt.xlabel('Number of Coin Throws (N)')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
def calculate_probability(target_step):
    # initial condition
    probabilities = [0] * (target_step + 1)
    probabilities[0] = 1  
    probabilities[1] = 0.5  #case of hit 2 from 1
    if target_step >= 2:
        probabilities[2] = 0.5 + 0.5*0.5 #case of hit 3 from 1

    # recurrence relationship
    for n in range(3, target_step + 1):
        probabilities[n] = 0.5 * (probabilities[n-1] + probabilities[n-2])

    #final answer
    return probabilities[target_step]


target_step = 25
probability = calculate_probability(target_step)
print(f"The probability of hitting step {target_step} is: {probability}")
```
